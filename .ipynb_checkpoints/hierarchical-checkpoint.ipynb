```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as hirar_utils
from sklearn.cluster import AgglomerativeClustering
import sklearn.metrics as metrics
from matplotlib_venn import venn2


sb.set_theme()

# import data frames
%store -r final_data
# simplified dataframe with Principal components
%store -r data6D
```

#Â Hierarchical model
We will compare four different linkage methods on euclidean distance

```python
data6DCP = data6D.copy()
countries = data6DCP["Country"]
data6DCP = data6DCP.drop(columns=["Country"])
plt.figure(figsize=(16,16))

# different types of linkages:
# ward - minizes variance
# average - uses the average of the distances of each observation of the two sets
# complete - uses the maximum distances between all observations of the two sets
# single - uses the minimum of the distances between all observations of the two sets

WARD = 'ward'
AVG = 'average'
COMPLETE = 'complete'
SINGLE = 'single'
DISTANCE = 'euclidean'
```

```python
# ward
plt.plot(figsize=(16,24))
plt.title(WARD + " distance")
dend = hirar_utils.dendrogram(hirar_utils.linkage(data6DCP, method=WARD, metric=DISTANCE))
```

```python
# average
plt.plot(figsize=(16,24))
plt.title(AVG + " distance")
dend = hirar_utils.dendrogram(hirar_utils.linkage(data6DCP, method=AVG, metric=DISTANCE))
```

```python
# complete
plt.plot(figsize=(16,24))
plt.title(COMPLETE + " distance")
dend = hirar_utils.dendrogram(hirar_utils.linkage(data6DCP, method=COMPLETE, metric=DISTANCE))
```

```python
# single
plt.plot(figsize=(50,50))
plt.title(SINGLE + " distance")
dend = hirar_utils.dendrogram(hirar_utils.linkage(data6DCP, method=SINGLE, metric=DISTANCE))
```

### Choosing a linkage
From the dendograms,  
1. ward and complete give us the most clear cluster information  
2. 3 clusters appears to be optimal for both linkages based on color code

```python
# do clustering
ward_model_3_clusters = AgglomerativeClustering(n_clusters=3, affinity=DISTANCE, linkage=WARD).fit_predict(data6DCP)
complete_model_3_clusters = AgglomerativeClustering(n_clusters=3, affinity=DISTANCE, linkage=COMPLETE).fit_predict(data6DCP)

# add cluster labels back to data
clustered_data = data6DCP.copy()
clustered_data["ward_model_3_clusters"] = pd.Categorical(ward_model_3_clusters)
clustered_data["complete_model_3_clusters"] = pd.Categorical(complete_model_3_clusters)


f, axes = plt.subplots(1,2, figsize=(16,8))
sb.countplot(x = "ward_model_3_clusters", data = clustered_data, ax=axes[0])
sb.countplot(x = "complete_model_3_clusters", data = clustered_data, ax=axes[1])
```

```python
# silhouette scores for the two models
ward_score = metrics.silhouette_score(data6DCP, ward_model_3_clusters, metric="euclidean")
complete_score = metrics.silhouette_score(data6DCP, complete_model_3_clusters, metric="euclidean")

print("ward model: ", ward_score)
print("complete model: ", complete_score)
```

### insight
Cluster 1 and 2 from the first ward model  
and cluster 2 from the complete model appear to contain the least developed countries

```python
ward_underdeveloped = clustered_data[clustered_data["ward_model_3_clusters"] != 0]["country"]
complete_underdeveloped = clustered_data[clustered_data["complete_model_3_clusters"] == 2]["country"]

union = pd.Series(np.union1d(ward_underdeveloped, complete_underdeveloped))
intersection = pd.Series(np.intersect1d(ward_underdeveloped, complete_underdeveloped))
notcommon = union[~union.isin(intersection)]

print("Countries that are not common to the clusters from the 2 models: ")
print(notcommon)
```

```python
plt.title("Ward model and complete model cluster overlaps")
venn2(subsets = (ward_underdeveloped.size, complete_underdeveloped.size, intersection.size), set_labels = ('Ward', 'Complete'))
```

# Further cluster interpretation
The complete model is not useful in distinguishing developing countries from under-developed countries.  
This is because one cluster contains all the non "first world countries"  

The ward model is more useful as it provides more distinction between developing countries in clusters 1 and 2.  

Therefore we will be keeping the ward model and comparing it with our insights from KMeans model

```python
#adding cluster to final_data
final_data_cp = final_data.copy()
final_data_cp = final_data_cp.reset_index(drop=True)
final_data_cp['cluster'] = clustered_data['ward_model_3_clusters']
```

```python
#   plotting boxplots of different clusters for the 15 variables in the final_data dataframe to study 
#   the actual distribution of the development indicators for each cluster (and for verifying that our 
#   clustering is effective)

for col in final_data_cp:
    if col == "country" or col == "Region" or col == "cluster":
        continue
    f = plt.figure(figsize=(18, 6))
    sb.boxplot(x = col, y = "cluster", data = pd.concat([final_data_cp[col], final_data_cp["cluster"]], axis = 1))
```
